```python
import numpy as np
from numpy.linalg import norm, det, inv, eig, matrix_rank
from sklearn.decomposition import PCA, TruncatedSVD
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
import numpy as np

s1 = np.array([88, 92, 85, 78, 95])
s2 = np.array([72, 68, 74, 85, 70])
s3 = np.array([95, 91, 89, 82, 98])
s4 = np.array([60, 55, 62, 70, 58])
s5 = np.array([78, 80, 75, 88, 83])
s6 = np.array([50, 48, 53, 65, 45])
```

## Part-A vector and matrix fundamentals

```python
np.linalg.norm(s1 , 1)
```

```python
np.dot(s1, s2)
```

```python
cos_theta = np.dot(s1, s2) / (np.linalg.norm(s1) * np.linalg.norm(s2))
angle = np.degrees(cos_theta)
angle
```

```python
#select 3 subjects
np.cross([88, 92, 85], [50, 48, 53])
```

```python
def projection(a, b):
    return (np.dot(a, b) / np.dot(b, b)) * b


proj_s1_on_s2 = projection(s1, s2)

print("Projection of s1 onto s2:")
print(proj_s1_on_s2)
```

## Part-B Matrix operations

```python
scores = np.array([
s1,
s2,
s3,
s4,
s5,
s6
])
```

```python
# Matrix Addition
scores + scores
```

```python

np.matmul(scores, scores.T)
```

```python
# transpose
scores.T
```

```python
#Determinant
A = scores[:5, :5]

np.linalg.det(A)
```

```python
A = scores[:5, :5]
np.linalg.inv(A)
```

## Part-D eigenvalues and decomposition

```python
cov = np.cov(scores.T)
cov
```

```python
eigenvalues, eigenvectors = np.linalg.eig(cov)

print(eigenvalues)
print(eigenvectors)
```

```python
U,S,Vt = np.linalg.svd(scores)

print(U)
print(S)
print(Vt)
```

## Part-E Dimensionality reduction

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(scores)

print(X_pca)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

labels = np.array([1,1,1,0,0,0])  # Above / Below average

lda = LinearDiscriminantAnalysis(n_components=1)
X_lda = lda.fit_transform(scores,labels)

print(X_lda)
```
