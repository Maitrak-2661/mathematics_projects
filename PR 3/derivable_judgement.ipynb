## Question-2

```python
def manual_CI(data,alpha):
    n = len(data)
    mean = np.mean(data)
    std = np.std(data,ddof=1)
    standard_error = std / np.sqrt(n)
    
    z_critical = stats.norm.ppf(1 - alpha /2)
    margin_error = z_critical * standard_error

    CI_upper = mean + margin_error
    CI_lower = mean - margin_error
    

    return CI_lower, CI_upper
```

```python
def scipy_CI(data,alpha):
    mean = np.mean(data)
    n = len(data)
    CI = stats.norm.interval(1 - alpha/2, loc=mean, scale=stats.sem(data))
    return CI
```

```python
cols_numerical= derive_judgements.select_dtypes(include="number").columns
```

```python
print("The CI using manual calculations")

for col in cols_numerical:
    CI = manual_CI(derive_judgements[col],alpha=0.05)
    print(f'{col.upper()} CI: ({CI[0]:.2f}, {CI[1]:.2f})')
```

```python
print("The CI using scipy Library")

for col in cols_numerical:
    CI = scipy_CI(derive_judgements[col],alpha=0.05)
    print(f'{col.upper()} CI: ({CI[0]:.2f}, {CI[1]:.2f})')
```

## Question-3

```python
def z_test(data,alpha):
    n = 35
    mu = np.mean(data)
    np.random.seed(18)
    s = np.random.choice(data, n)
    sigma = np.std(data)
    x_bar = np.mean(s)
    alpha = alpha
    
    z_score = (x_bar - mu) / (sigma / np.sqrt(n))
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    critical_z = stats.norm.ppf(1 - alpha/ 2)
    

    return z_score,p_value,critical_z
```

```python
for col in cols_numerical:
    z = z_test(derive_judgements[col],alpha=0.05)
    print(f'\n{col.upper()}: ')

    print(f'z_score: {z[0]}')
    print(f'p_value: {z[1]}')
    print(f'critical_z: {z[2]}')
```

```python
def t_test(data,alpha):
    n = 25
    mu = np.mean(data)
    np.random.seed(18)
    t = np.random.choice(data, n)
    S = np.std(t)
    x_bar = np.mean(t)
    alpha = alpha
    
    t_score = (x_bar - mu) / (S/ np.sqrt(n))
    
    p_value = 2 * (1 - stats.t.cdf(abs(t_score), df = n-1))

    critical_t = stats.t.ppf(1 - alpha/2, df= n-1)

    return t_score, p_value, critical_t
```

```python
for col in cols_numerical:
    t = t_test(derive_judgements[col],alpha=0.05)
    print(f'\n{col.upper()}: ')

    print(f't_score: {t[0]}')
    print(f'p_value: {t[1]}')
    print(f'critical_t: {t[2]}')
```

```python
n1 = len(group1)
print(n1)
n2 = len(group2)
print(n2)

mean1 = np.mean(group1)
mean2 = np.mean(group2)
```

```python
print("Group 1 (Smokers) =", n1, " Mean BMI =", mean1)
print("Group 2 (Non-Smokers) =", n2, " Mean BMI =", mean2)
```

```python
alpha = 0.05

if n1 >= 30 and n2 >= 30:
    print("Using Z-test as the sample size is large")
    
    S1 = np.std(group1, ddof=1)
    S2 = np.std(group2, ddof=1)
    
    z_score = (mean1 - mean2) / np.sqrt((S1**2 / n1) + (S2**2 / n2))
    z_critical = stats.norm.ppf(1 - alpha/2)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    
    print("z_score:", z_score)
    print("z_critical:", z_critical)
    print("p_value:", p_value)
    
    print("\nconclusion with z_score")
    if z_score < -z_critical or z_score > z_critical:
        print("we reject H0, as there is significant difference between mean bmi of smoker and mean bmi of non-smoker")
    else:
        print("we fail to reject H0, as there is no significant difference between mean bmi of smoker and mean bmi of non-smoker")
    
    print("\nconclusion with p_value")    
    if p_value < alpha:
        print("we reject H0, as there is significant difference between mean bmi of smoker and mean bmi of non-smoker")
    else:
        print("we fail to reject H0, as there is no significant difference between mean bmi of smoker and mean bmi of non-smoker")
else:
    print("using t-test as the sample size is small") 
    t_score, p_value = stats.ttest_ind(group1, group2, equal_var=False)
    t_critical = stats.t.ppf(1 - alpha/2, df=n1 + n2 - 2)
    
    print("t_score:", t_score)
    print("t_critical:", t_critical)
    print("p_value:", p_value)
    
    
    print("\nconclusion with t_score")
    if t_score < -t_critical or t_score > t_critical:
        print("we reject H0, as there is significant difference between mean bmi of smoker and mean bmi of non-smoker")
    else:
        print("we fail to reject H0, as there is no significant difference between mean bmi of smoker and mean bmi of non-smoker")
      
    print("\nconclusion with p_value")  
    if p_value < alpha:
        print("we reject H0, as there is significant difference between mean bmi of smoker and mean bmi of non-smoker")
    else:
        print("we fail to reject H0, as there is no significant difference between mean bmi of smoker and mean bmi of non-smoker")
      
```

```python
from scipy.stats import chi2
H0 = 'smoking_status and diabetes are independent categorical variables'
H1 = 'smoking_status and diabetes are dependent categorical variables'

observed = pd.crosstab(derive_judgements['smoking_status'], derive_judgements['diabetes']).values
print(observed)

row_totals = observed.sum(axis=1)
col_totals = observed.sum(axis=0)
grand_total = observed.sum()

expected = np.outer(row_totals,col_totals) / grand_total

chi_score= ((observed - expected) ** 2 / expected).sum()

rows, cols = observed.shape
df = (rows - 1) * (cols - 1)

alpha = 0.05

p_value = 1 - stats.chi2.cdf(chi_score, df)
critical_chi = stats.chi2.ppf(1 - alpha, df)

print("chi_score:", chi_score)
print("critical_chi:", critical_chi)
print("p_value:", p_value)

print("\nconclusion with chi_score")
if chi_score > critical_chi:
    print("we reject H0, as smoking status and diabetes is dependent")
else:
    print("we fail to reject H0, as smoking status and diabetes are independent")
    
print("\nconclusion with p_value")  
if p_value < alpha:
    print("we reject H0, as smoking status and diabetes is dependent")
else:
    print("we fail to reject H0, as smoking status and diabetes are independent")
```

## Question 6

```python
from scipy.stats import f_oneway

H0 = 'Disease rate is same across age_group'
H1 = 'Disease rate differ with age_group'

derive_judgements['age_group'] = pd.cut(derive_judgements['age'], bins=[0, 29, 49, 100], labels=['Young', 'Middle', 'Senior'])


derive_judgements['disease'] = (
    derive_judgements['diabetes']
    .map({'True': 1, 'False': 0, True: 1, False: 0})
    .astype('Int64')
)


young = derive_judgements[derive_judgements['age_group'] == 'Young']['disease']
middle = derive_judgements[derive_judgements['age_group'] == 'Middle']['disease']
senior = derive_judgements[derive_judgements['age_group'] == 'Senior']['disease']

groups = [young,middle,senior]

all_data = np.concatenate(groups)
grand_mean = np.mean(all_data)

SSB = sum(len(g) * (np.mean(g) - grand_mean)**2 for g in groups)
SSW = sum(sum((g - np.mean(g))**2) for g in groups)

k = len(groups)
N = len(all_data)

df_between = k - 1
df_within = N - k

MSB = SSB / df_between
MSW = SSW / df_within

F_statistic = MSB / MSW

print(f"f_stats:-{F_statistic}")

F_auto, p_value = stats.f_oneway(young,middle,senior)
alpha = 0.05
critical_F = stats.f.ppf(1 - alpha, df_between, df_within)

print(f"f_auto:-{F_auto}") 
print(f"p_value:-{p_value}")
print(f"critical_F:-{critical_F}")

print("\nconclusion with f_stats")
if F_statistic > critical_F:
    print("we reject H0, as disease rate differ with age group")
else:
    print("we fail to reject H0, as disease rate is same across age_group")
    
print("\nconclusion with p_value")  
if p_value < alpha:
    print("we reject H0, as disease rate differ with age group")
else:
    print("we fail to reject H0, as disease rate is same across age_group")
```
